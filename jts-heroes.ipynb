## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#blah
drop_name_df = purchase_data.drop_duplicates(subset = ["SN"])


#blah
males = (drop_name_df["Gender"].value_counts().Male)

females = (drop_name_df["Gender"].value_counts().Female)

others = len(drop_name_df.loc[drop_name_df["Gender"] == "Other / Non-Disclosed",:])

#blah
percentmales = (males/(males+females+others))*100

percentfemales = (females/(males+females+others))*100

percentothers = (others/(males+females+others))*100

total_players = (females+males+others)


#blah
GenderDemo_df = pd.DataFrame({"Total Count":[males, females, others],
                                      "Percentage of Players":[percentmales, percentfemales, percentothers]},
                                      index =["Male", "Female", "Other/Non-Disclosed"])

GenderDemo_df["Percentage of Players"] = GenderDemo_df["Percentage of Players"].map("{:.2f}%".format)                             

GenderDemo_df
                                                                                    
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
#reset index using gender
male_df = purchase_data.loc[purchase_data["Gender"] == "Male",:]

#blah
malepc = male_df.shape[0]

maleapp = male_df["Price"].mean()

maletpv = male_df["Price"].sum()

maleatppp = maletpv/males


female_df = purchase_data.loc[purchase_data["Gender"] == "Female",:]

femalepc = female_df.shape[0]

femaleapp = female_df["Price"].mean()

femaletpv = female_df["Price"].sum()

femaleatppp = femaletpv/females



other_df = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed",:]

otherpc = other_df.shape[0]

otherapp = other_df["Price"].mean()

othertpv = other_df["Price"].sum()

otheratppp = othertpv/others


pag_df = pd.DataFrame({"Purchase Count":[femalepc, malepc, otherpc],
                                      "Average Purchase Price":[femaleapp, maleapp, otherapp],
                                       "Total Purchase Value":[femaletpv, maletpv, othertpv],
                                       "Avg Total Purchase per Person":[femaleatppp, maleatppp, otheratppp]},
                                       index =["Female", "Male", "Other/Non-Disclosed"])
                              

pag_df["Average Purchase Price"] = pag_df["Average Purchase Price"].map("${:.2f}".format)
pag_df["Total Purchase Value"] = pag_df["Total Purchase Value"].map("${:.2f}".format)
pag_df["Avg Total Purchase per Person"] = pag_df["Avg Total Purchase per Person"].map("${:.2f}".format)

pag_df.head()

```
